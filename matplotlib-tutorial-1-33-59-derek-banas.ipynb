### Functional Plots

```python
x_1 = np.linspace(start=0, stop=5, num=10)
y_1 = x_1**2
# print(x_1)
# print(y_1)
# print(type(x_1))
# print(type(y_1))
plt.title('Days Squared Chart')
plt.xlabel('Days')
plt.ylabel('Days Squared')
plt.plot(x_1, y_1, c="b")
plt.show()
```

### Print Multiple Plots

```python
plt.subplot(1, 2, 1)
plt.plot(x_1, y_1, 'r')

plt.subplot(1, 2, 2)
plt.plot(x_1, y_1, 'b')

plt.show()
```

### Using figure object
- A `figure` is an object that contains all the plot elements and can contain many axes

```python
fig_1 = plt.figure(figsize=(5, 4), dpi=100)
# left, bottom, width, height
axes_1 = fig_1.add_axes([0.1, 0.1, 0.9, 0.9])
axes_1.set_xlabel('Days')
axes_1.set_ylabel('Days Squared')
axes_1.set_title('Days Squared Chart')
axes_1.plot(x_1, y_1, label='x/x^2')
axes_1.plot(y_1, x_1, label='x^2/x')

# upper right: 1, upper left: 2, lower left: 3, lower right: 4
# or supply a tuple of x & y from lower left
# axes_1.legend(loc=0)
axes_1.legend(loc=0)

axes_2 = fig_1.add_axes([0.45, 0.45, 0.4, 0.3])
axes_2.set_xlabel('Days')
axes_2.set_ylabel('Days Squared')
axes_2.set_title('Days Squared Chart')
axes_2.plot(x_1, y_1, 'r')
axes_2.text(0, 40, 'Message')
plt.show()
```

### SubPlots

```python
fig_2, axes_2 = plt.subplots(figsize=(12, 4), nrows=1, ncols=3)
plt.tight_layout()
axes_2[0].set_title('Plot 1')
axes_2[1].set_title('Plot 2')
axes_2[2].set_title('Plot 3')
axes_2[0].set_xlabel('X')
axes_2[0].set_ylabel('X Squared')
axes_2[1].set_xlabel('X')
axes_2[1].set_ylabel('X Squared')
axes_2[2].set_xlabel('X')
axes_2[2].set_ylabel('X Squared')
axes_2[0].plot(x_1, y_1, 'burlywood')
axes_2[1].plot(x_1, y_1, 'r')
axes_2[2].plot(x_1, y_1, 'y')

plt.show()
```

### Appearance Options
- Default colors: ('b': blue, 'g': green, 'r': red, 'y': yellow, 'c': cyan, 'm': magenta, 'k': black, 'w': white)
- `color="0.75"` creates 75% gray
- We can use hexcodes for colors, eg, `color="#eeefff"`
- We can also use color names like `color="burlywood"`
- [Web Colors](https://en.wikipedia.org/wiki/Web_colors)

```python
fig_3 = plt.figure(figsize=(10, 4))
axes_3 = fig_3.add_axes([0, 0, 1, 1])
axes_3.set_title('X Squared Chart')
axes_3.set_xlabel('X')
axes_3.set_ylabel('X Squared')
# `alpha` represents the transparency, `lw` stands for line width which is 1 (by default), `ls` means line style
axes_3.plot(x_1, y_1, color='navy', alpha=0.95, lw=2, ls='--', marker='x', markersize=7, markerfacecolor='k', markeredgecolor='r', markeredgewidth=5, label='Plot 1')
# axes_3.plot(y_1, x_1, 'k', label='Plot 2')
axes_3.set_xlim([0, 3])
axes_3.set_ylim([0, 8])
axes_3.grid(True, color='navy', dashes=(5, 2, 1, 2))
axes_3.set_facecolor("#ff34a9")

plt.show()
```

### Working with Pandas DataFrame

```python
icecream_sales_df = pd.read_csv('./matplotlib-tutorial-data/icecreamsales.csv')
icecream_sales_df = icecream_sales_df.sort_values(by='Temperature')
icecream_sales_df
```

Converting the pandas DataFrame to numpy ndarray

```python
np_arr = icecream_sales_df.values
print(np_arr)
print(np_arr.shape)
print(len(np_arr))
print(np_arr[-1][-1])
```

```python
x = np_arr[:, 0]
print('Shape of x:', x.shape, 'Type of x:', type(x), 'x:', x)
y = np_arr[:, 1]
print('Shape of y:', y.shape, 'Type of y:', type(y), 'y:', y)
```

```python
fig = plt.figure(figsize=(12, 6))
axes1 = fig.add_axes([0.05, 0.05, 0.4, 0.4])
axes2 = fig.add_axes([0.55, 0.55, 0.4, 0.4])
axes3 = fig.add_axes([0.55, 0.05, 0.4, 0.4])
axes4 = fig.add_axes([0.05, 0.55, 0.4, 0.4])
axes1.plot(x, y, label='axes1')
axes1.set_title('axes1')
axes2.plot(x, y, label='axes2', color='r')
axes2.set_title('axes2')
axes2.set_xlabel('Temperature')
axes2.set_ylabel('Ice cream Sales')
axes2.annotate('Good month!', xy=(81, 556), xytext=(55, 525), arrowprops=dict(facecolor='black', shrink=0.01))
axes3.plot(x, y, label='axes3', color='y')
axes3.set_title('axes3')
axes4.plot(x, y, label='axes4', color='navy')
axes4.set_title('axes4')
axes2.bar(x, y)
plt.show()
```

### Histograms
- `cumulative=True` to use a cumulative distribution
- `histtype='step'` generates a line plot
- `color='orange'` to change the color of the plot
- `orientation='horizontal'`
- Range gets a tuple with bin range interested in

```python
arr_1 = np.random.randint(1, 7, 5000)
arr_2 = np.random.randint(1, 7, 5000)
print(type(arr_1))
print(arr_1.shape)
print(arr_1)
print(type(arr_2))
print(arr_2.shape)
print(arr_2)
print(arr_1 == arr_2)
```

```python
arr_1 = np.random.randint(1, 7, 1000000)
arr_2 = np.random.randint(1, 7, 1000000)
arr_3 = arr_1 + arr_2

plt.hist(arr_3, bins=11, stacked=True, density=True, color='orange', cumulative=False, histtype='bar', orientation='vertical')
plt.xlabel('Dice Sum')
plt.ylabel('Probability')
plt.xticks(range(2, 13))
plt.show()
```
